```python
import numpy as np
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
def calculos (amostras):
    # Cálculo da Média (µ)
    media = np.mean(amostras)
    # Variância (Var)
    variancia = np.var(amostras, ddof=1)  # usando n-1 para estimativa amostral
    # Desvio Padrão (DP)
    desvio_padrao = np.sqrt(variancia)
    # Amplitude (A)
    amplitude = np.max(amostras) - np.min(amostras)
    # Quartis e Amplitude Interquartil (AIQ)
    Q1 = np.percentile(amostras, 25)
    Q3 = np.percentile(amostras, 75)
    amplitude_interquartil = Q3 - Q1
    # Coeficiente de Variação (CV)
    coeficiente_variacao = (desvio_padrao / media) * 100
    # Curtose (K)
    curtose = stats.kurtosis(amostras, fisher=True)
    # Exibindo os Resultados
    print(f"Média (µ): {media:.2f}")
    print(f"Variância (Var): {variancia:.2f}")
    print(f"Desvio Padrão (DP): {desvio_padrao:.2f}")
    print(f"Amplitude (A): {amplitude:.2f}")
    print(f"Amplitude Interquartil (AIQ): {amplitude_interquartil:.2f}")
    print(f"Coeficiente de Variação (CV): {coeficiente_variacao:.2f}%")
    print(f"Curtose (K): {curtose:.2f}")
```

```python
def grafico(amostras):
     # Plotar o histograma das amostras
    plt.hist(amostras, bins=30, density=True, alpha=0.6, color='g', edgecolor='black')
    # Ajustar uma distribuição normal e plotar o histfit
    mu, std = norm.fit(amostras)  # Ajustar os parâmetros da distribuição normal
    # Criar uma faixa de valores para plotar a curva de ajuste
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    p = norm.pdf(x, mu, std)
    # Plotar a curva de ajuste
    plt.plot(x, p, 'k', linewidth=2)
    # Adicionar título e rótulos aos eixos
    plt.title('Histograma com Ajuste de Distribuição Normal')
    plt.xlabel('Valores')
    plt.ylabel('Densidade')
    # Mostrar o gráfico
    plt.show()
```

```python
n_amostras=100000
media_populacional=2
desvio_padrao_populacional=10
# Gerando n amostras de uma distribuição normal
amostras = np.random.normal(loc=media_populacional, scale=desvio_padrao_populacional, size=n_amostras)
# Amostra 1: Distribuição Exponencial
amostras_exponenciais = np.random.exponential(scale=1, size=n_amostras)
# Amostra 2: Distribuição Uniforme
amostras_uniformes = np.random.uniform(low=0, high=1, size=n_amostras)
```

```python
print("Variáveis Aleatórias Normais")
calculos(amostras)
print("\nVariáveis Aleatórias Exponenciais")
calculos(amostras_exponenciais)
print("\nVariáveis Aleatórias Uniformes")
calculos(amostras_uniformes)
```

```python
grafico(amostras)
grafico(amostras_exponenciais)
grafico(amostras_uniformes)
```
