```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.mosaicplot import mosaic
```

```python
# Gerar dados fictícios para o exemplo
n= 500
#np.random.seed(42)
generos = np.random.choice(['Masculino', 'Feminino'], size=n)
preferencias = np.random.choice(['Cinema', 'Esporte', 'Música'], size=n)
```

```python
# Criar DataFrame
dados = pd.DataFrame({'Gênero': generos, 'Preferência de Lazer': preferencias})
dados
```

```python
# Construir a tabela de contingência
tabela_contingencia = pd.crosstab(dados['Gênero'], dados['Preferência de Lazer'], margins=True, margins_name="Total")
tabela_contingencia
```

```python
# Preparar dados no formato de dicionário para o gráfico de mosaico
dados_mosaico = {(row, col): tabela_contingencia.loc[row, col]
                 for row in ['Feminino', 'Masculino']
                 for col in ['Cinema', 'Esporte', 'Música']}
```

```python
# Criar gráfico de mosaico
fig, ax = plt.subplots(figsize=(8,6))
mosaic(dados_mosaico, title="Gráfico de Mosaico - Gênero x Preferência de Lazer", ax=ax)

plt.show()
```

```python
dados.to_csv('dados.csv', index=False)
#
```
